# Branching and Merging

Creating separate lines of development and integrating changes safely.

## 1. git branch

**Theory**

Branching is one of Git's most powerful features. A branch represents an independent line of development. When you want to work on a new feature or fix a bug, you spawn a new branch to encapsulate your changes. This ensures that unstable code does not affect the main codebase.

Technically, a branch in Git is essentially a lightweight movable pointer to a commit. The default branch name in Git is usually `main` or `master`.

**Commands**
```bash
# List all local branches (the current branch is highlighted with an asterisk)
git branch

# Create a new branch, but stay on the current branch
git branch feature-branch-name

# Delete a branch (safe delete: prevents deleting unmerged changes)
git branch -d branch-to-delete
```

## 2. git checkout / git switch

**Theory**

To start working on a different branch, you need to switch to it. This updates the files in your working directory to match the version stored in that branch. 

Traditionally, `git checkout` is used for this. In newer versions of Git, `git switch` was introduced as a more explicit command for switching branches.

**Commands**

```bash
# Switch to an existing branch
git checkout feature-branch-name
# OR
git switch feature-branch-name

# Create a new branch and switch to it immediately
git checkout -b new-feature
# OR
git switch -c new-feature
```

## 3. git merge

**Theory**

Merging is the way you integrate changes from one branch into another. Common patterns include merging a completed feature branch back into the main branch.

When you merge, Git identifies the common base commit between the two branches and creates a new "merge commit" that combines the changes. If there are conflicting changes (e.g., the same line modified differently in both branches), Git will pause and ask you to resolve the conflict manually.

**Commands**
```bash
# Merge the source branch into the current branch
git merge feature-branch-name

# Abort a merge in case of conflicts or error
git merge --abort
```

## Practice Tasks

### Task 1:Create a first branch

```bash
# In the previously created repository, create a new branch named "feature-branch"
git branch feature-branch

# Switch to the new branch
git switch feature-branch

# List all branches
git branch
```

### Task 2: Merge changes made on feature-branch to main branch

```bash
# Add another file in the new branch
touch newfile.txt

# Check the status of the new file
git status

# Add the new file to the staging area
git add newfile.txt

# Check the status of the new file
git status

# Commit the new file
git commit -m "Add newfile.txt"

# Switch back to the main branch
git switch main

# Merge the new branch into the main branch
git merge feature-branch

# Push the changes to the upstream that was previously set
git push
```

### Task 3: Add upstream for the new branch and push it to the remote repository

Now currently on the remote repository only the main branch exists. 

Let's set the upstream for the new branch also and push it to the remote repository.

```bash
# switch to the feature branch
git switch feature-branch

# This time let us not use -u to perform setting the upstream branch and pushing in the same command
# Let us use two separate commands for each step
# 1. set upstrean for the current branch
git --set-upstream-to=origin/feature-branch feature-branch
# 2. now push the branch to the set upstream
git push
```
